## Basic Charting

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [2,5,7])
plt.show()
```

```python
# adding titles and labels
plt.plot([1,2,3], [2,5,7])
plt.title("Hammad's Data")
plt.xlabel("This is x")
plt.ylabel("This is y")
plt.show()
```

```python
# two different lines in same plot
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,3]
y2 = [5,8,9]

plt.plot(x,y, label='first', linewidth=3)
plt.plot(x2,y2, label='second')

plt.legend()    # label ko show karne ke liye
plt.show()
```

```python
plt.scatter(x2,y2, label='second')
```

```python
plt.hist(x2,y2, label='second')
```
